```python
import time

def job_detail(url):
    # Return dict info of Job
    resp = requests.get(url)
    info = json.loads(resp.text)
    return info

start = time.time()

# Extract data and write to csv file
with open('job_draft.csv', 'w') as file:
    f = csv.writer(file)
    f.writerow(keys)
    for i in range(1, 1496): # id from 1 to 1495
        url = 'http://jobs.pymi.vn/jobs/{0}/Tester/?api=True'.format(i)
        try:
            detail = [job_detail(url)[k] for k in keys] # list of job's detail
        except:
            pass
        f.writerow(detail)

# Calculation duration time
duration = time.time() - start
print('Took %s seconds' % duration)
```
